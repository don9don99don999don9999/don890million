```python
df2 = df.groupby(["source","target"]).size().reset_index()
```

```python
unique_gene = pd.Index(df2['source']
                      .append(df2['target'])
                      .reset_index(drop=True).unique())
```

```python
df2
```

```python
unique_gene[:5]
```

```python
group_dict1 = {}
counter = 0
for gene in unique_gene:
    what_gene = re.match("^(\D)(\D)(\d)(\D)(\d)(\d)(\d)(\d)(\d)$", gene)
    if what_gene:
        net_gene = ''.join(what_gene.group(0))
        if net_gene not in group_dict1:
            counter += 1
            group_dict1[net_gene] = counter
        else:
            pass
```

```python
df2.rename(columns={0:'count'}, inplace=True)
temp_links_list = list(df2.apply(lambda row: {"source": row['source'], "target": row['target'], "value": row['count']}, axis=1))
```

```python
links_list = []
for link in temp_links_list:
    record = {"value":link['value'], "source":unique_gene.get_loc(link['source']),
     "target": unique_gene.get_loc(link['target'])}
    links_list.append(record)
```

```python
nodes_list = []

for gene in unique_gene:
    what_gene = re.match("^(\D)(\D)(\d)(\D)(\d)(\d)(\d)(\d)(\d)$", gene)
    if what_gene:
        net_gene = ''.join(what_gene.group(0))
        nodes_list.append({"name":gene, "group": group_dict1.get(net_gene)})
```
