```python


import pandas as pd

import json
import re
import warnings
```

```python
df = pd.read_csv('/home/plantgenomics/WDU/downloads/AraNet.csv',sep='\t') #아라넷 ->DF
```

```python
df2 = df.groupby(["source","target"]).size().reset_index() #source,target만 가져옴.
```

```python
df100=df
```

```python
df100.head()
```

```python
selected_genelist=['AT1G73590', 'AT2G14960', 'AT2G39370', 'AT3G19200'] #예진이가 만들어준 옥신관련유전자4개
```

```python
df100['source_test'] = df.apply(lambda x : x['source'] in set(selected_genelist),axis=1) #4개포함된것 찾
df100['target_test'] = df.apply(lambda x : x['target'] in set(selected_genelist),axis=1) #4개포함된것 찾
```

```python
mm1 = df100['source_test'] == False #마아아스킹
mm2 = df100['target_test'] == False #마아아스킹
mm = mm1&mm2 #마아아스킹 &마아아스킹

```

```python
df100[~mm] #df100에서 마스킹.
```

```python
df200 = df100[~mm].groupby(["source","target"]).size().reset_index() #source,target항목만 df200으로 만듬
```

```python
udu = list(set(list(df100[~mm]['source']) + list(df100[~mm]['target'])))
```

```python
group_dict100 = {}
counter = 0
for gene in udu:
    what_gene = re.match("^(\D)(\D)(\d)(\D)(\d)(\d)(\d)(\d)(\d)$", gene)
    if what_gene:
        net_gene = ''.join(what_gene.group(0))
        if net_gene not in group_dict100:
            counter += 1
            group_dict100[net_gene] = counter
        else:
            pass
```

```python
nodes_list200 = []

for gene in udu:
    what_gene2 = re.match("^(\D)(\D)(\d)(\D)(\d)(\d)(\d)(\d)(\d)$", gene)
    if what_gene2:
        net_gene2 = ''.join(what_gene2.group(0))
        nodes_list200.append({"name":gene, "group": group_dict100.get(net_gene2)})
```

```python
unique_gen = pd.Index(df200['source']
                      .append(df200['target'])
                      .reset_index(drop=True).unique()) #source,target에서 중복없는 유전자 찾음
```

```python
df200.rename(columns={0:'count'}, inplace=True)
temp_links_list200 = list(df200.apply(lambda row: {"source": row['source'], "target": row['target'], "value": row['count']}, axis=1))
```

```python
links_list200 = []
for link in temp_links_list200:
    record = {"value":link['value'], "source":unique_gen.get_loc(link['source']),
     "target": unique_gen.get_loc(link['target'])}
    links_list200.append(record)
```

```python

json_prep200= {"links":links_list200, "nodes":nodes_list200}

json_prep200.keys()
#json 형식으로 만들어버리기
```

```python
json_dump=json.dumps(json_prep200,indent=1, sort_keys=True)
```

```python
filename_out='exp.json' #.json 파일 생성
json_out = open(filename_out,'w')
json_out.write(json_dump)
json_out.close()
```
